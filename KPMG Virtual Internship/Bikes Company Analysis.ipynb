```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path_to_file = '../Projects/Bikes Company.xlsx'
```

```python
xl = pd.ExcelFile(path_to_file)
xl.sheet_names
```

```python
pd.options.display.max_columns = None
pd.options.display.max_colwidth = 100
```

```python
xl1 = xl.parse('Transactions')
```

```python
display(xl1.head(), xl1.tail(), xl1.dtypes, xl1.shape)
```

```python
xl1['transaction_id'] = xl1['transaction_id'].astype(str)
xl1['product_id'] = xl1['product_id'].astype(str)
xl1['customer_id'] = xl1['customer_id'].astype(str)
```

```python
xl1.isna().sum()
```

```python
xl1['standard_cost'].fillna(0, inplace=True)
```

```python
xl1 = xl1.fillna('NA')
```

```python
xl1['profit'] = xl1['list_price'] - xl1['standard_cost']
```

```python
xl2 = xl.parse('Customer Demographic')
```

```python
display(xl2.head(), xl2.tail(), xl2.dtypes, xl2.shape)
```

```python
xl2['customer_id'] = xl2['customer_id'].astype(str)
xl2['Age Group'] = xl2['Age Group'].astype(str)
```

```python
xl2.isna().sum()
```

```python
xl2 = xl2.fillna('NA')
```

```python
xl3 = xl.parse('Customer Address')
```

```python
display(xl3.head(), xl3.tail(), xl3.dtypes, xl3.shape)
```

```python
xl3['postcode'] = xl3['postcode'].astype(str)
xl3['customer_id'] = xl3['customer_id'].astype(str)
```

```python
left_join_result = pd.merge(xl1, xl2, on='customer_id', how='left')
final = pd.merge(left_join_result, xl3, on='customer_id', how='left')
final
```

```python
final.isna().sum()
```

```python
final = final.fillna('NA')
```

```python
final['online_order'].value_counts()
```

```python
# Define the Excel writer
with pd.ExcelWriter('Bikes Full.xlsx', engine='xlsxwriter') as writer:
    # Write each DataFrame to a separate sheet in the Excel file
    final.to_excel(writer, sheet_name='Sheet1', index=False)
```
