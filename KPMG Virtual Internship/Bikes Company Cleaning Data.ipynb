```python
# Function to calculate age based on the current date
def calculate_age(birthdate):
    today = datetime.today()
    age = today.year - birthdate.year - ((today.month, today.day) < (birthdate.month, birthdate.day))
    return age

# Calculate and update the ages
xl2['Age'] = xl2['DOB'].apply(calculate_age)
```
